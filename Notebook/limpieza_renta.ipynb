```python
import pandas as pd

df = pd.read_csv('panel_indicadores_distritos_barrios.csv', sep=';', encoding='utf-8')

df
```

```python
renta = df[
    (df['indicador_nivel1'].str.strip() == 'Renta') &
    (df['indicador_nivel2'].str.strip() == 'Bruta Per cápita') &
    (df['indicador_completo'].str.strip() == 'Renta disponible media por persona') &
    (df['categoría_1'].str.strip() == 'Indicadores Económicos')
]
```

```python
columnas_a_eliminar = [
    'fuente_indicador', 'categoría_1', 'categoría_2',
    'indicador_nivel3', 'unidad_indicador', 'ciudad',
    'Periodo panel', 'año', 'fecha_indicador', 'indicador_nivel1',
    'indicador_nivel2', 'indicador_completo', 'Orden'
]

renta = renta.drop(columns=columnas_a_eliminar)
```

```python
renta = renta.dropna(subset=['barrio'])
```

```python
renta['barrio'] = renta['barrio'].str.strip()
```

```python
renta['cod_barrio'] = renta['cod_barrio'].astype(int)
```

```python

renta['valor_indicador'] = renta['valor_indicador'].astype('int64')
```

```python
renta.to_csv('12_renta_2024.csv', index=False)
```
